<h2>Use Jieba 結巴</h2>

```python
import jieba
lstWord = ['我要成為香港神奇寶寶大師', '我家門前有條水溝很難過', '電視主播呂藝仁曬婚紗照報喜']
# 全模式
Sentence1=jieba.cut(lstWord[0], cut_all=True)
Sentence2=jieba.cut(lstWord[1], cut_all=True)
Sentence3=jieba.cut(lstWord[2], cut_all=True)
# 精準模式
Sentence_1=jieba.cut(lstWord[0], cut_all=False)
Sentence_2=jieba.cut(lstWord[1], cut_all=False)
Sentence_3=jieba.cut(lstWord[2],cut_all=False)

print('1A. ', '全模式'+":"+"/".join(Sentence1))
print('1B. ', '精準模式'+":"+"/".join(Sentence_1))
print('2A. ', '全模式'+":"+"/".join(Sentence2))
print('2B. ', '精準模式'+":"+"/".join(Sentence_2))
print('3A. ', '全模式'+":"+"/".join(Sentence3))
print('3B. ', '精準模式'+":"+"/".join(Sentence_3))
```

<h2>JieBa 詞性標籤</h2>

```python
# importing the module
import json

json_file = open('.//Material05//JieBaTag.json', 'r', encoding='utf-8')
dictTag = json.load(json_file)

def findWordName(_key: str) -> str:
    if _key in dictTag:
        return dictTag[_key]
    else:
        return 'None'
```

```python
print('關於詞性')
import jieba.posseg as pseg
Sentence1=pseg.cut('我要成為香港神奇寶寶大師')
Sentence2=pseg.cut('我家門前有條水溝很難過')
Sentence3=pseg.cut('電視主播呂藝仁曬婚紗照報喜')
for word,flag in Sentence1:
  print(word, flag, findWordName(flag)[0])
print('-' * 70)
for word, flag in Sentence2:
  print(word, flag, findWordName(flag)[0])
print('-' * 70)
for word, flag in Sentence3:
  print(word, flag, findWordName(flag)[0])
print('-' * 70)
```

```python
import jieba

with open('.//Material05//data.txt', 'r', encoding='UTF-8') as file:
    text = file.read()
    # print(text)
    # print("*"*70)
    segments = jieba.cut(text)
    print('type:', type(segments))
    lstSegment = []
    for seg in segments:
        #print(seg)        
        #if len(seg) == 1: # exclude the single character
        #    continue
        lstSegment.append(seg)
    print(lstSegment)
```

```python
# Top 10 words in the article
import collections
counter1 = collections.Counter(lstSegment)
print("顯示 Top 10 詞彙：")
for key, val in counter1.most_common(10):
    print(key, ': ', val)
```

```python
import jieba

segments = []
stopwords = []
remainderwords = []
lstSegment_2 = []

with open('.//Material05//stopWords.txt', 'r', encoding='UTF-8') as file:
    for data in file:
        stopwords.append(data.strip())

with open('.//Material05//data.txt', 'r', encoding='UTF-8') as file:
    text = file.read()
    segments = jieba.cut(text)

# fill the remainderwords
remainderwords = list(filter(lambda x: x not in stopwords and x != '\n', segments))

#remainderwords = [x for x in segments if (x not in stopwords and x != '\n')]

for remd in remainderwords:
    lstSegment_2.append(remd)
print(lstSegment_2)
```

```python
# Top 10 words in the article
# not including stop words
import collections
counter2 = collections.Counter(remainderwords)
print("顯示 Top 10 詞彙：")
for key, val in counter2.most_common(10):
    print(key, ': ', val)
```

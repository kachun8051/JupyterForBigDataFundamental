```python
import camelot
import openpyxl

file_path = './full_list_chi_1.pdf'

# Read the PDF file and extract the tables
tables = camelot.read_pdf(file_path, pages='all')

# Create a new Excel workbook and worksheet
workbook = openpyxl.Workbook()
worksheet = workbook.active

# Loop through the extracted tables and write them to the Excel worksheet
for i, table in enumerate(tables, start=1):
    # Write the table data to the worksheet
    for row in table.data:
        worksheet.append(row)

    # Add a sheet title
    worksheet.title = f"Table {i}"

    # Create a new worksheet for the next table
    if i < len(tables):
        worksheet = workbook.create_sheet(f"Table {i+1}")

# Save the Excel workbook
workbook.save('../output04/output.xlsx')
print('done!')
```

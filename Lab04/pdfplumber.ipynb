<h2>Define the class of row</h2>

```python
class clsRow:
    def __init__(self):
        self.clearIt()
    
    def getRowNum(): 
        return self.rownum

    def getName():
        return self.name

    def getAddress():
        return self.address

    def getArea():
        return self.area
    
    def clearIt(self):
        self.rownum = ''
        self.name = ''
        self.address = ''
        self.area = ''

    def isEmpty(self) -> bool:
        if self.rownum == '':
            return True
        else:
            return False

    @staticmethod
    def myNewLine(x: str, y: str) -> str:
        if x == '':
            return y
        if y == '':
            return x
        return x + '\n' + y
    
    def addRow(self, _row: str, _name: str, _addr: str, _area: str) -> bool:
        if self.rownum == '': # new assign
            if _row == '':
                print(f"Impossible case happened!!!\nName:{_name}")
                return False
            else:
                self.rownum = _row
                self.name = _name
                self.address = _addr
                self.area = _area
                # print(_name, ' is added first time.')
                return True
        else: 
            if _row == '': # next assign
                self.name = self.myNewLine(self.name, _name)
                self.address = self.myNewLine(self.address, _addr)
                self.area = self.myNewLine(self.area, _area)
                # print(_name, ' is added 1+ time.')
                return True
            else:
                # print(_name, ' is not added!')
                return False

    def isRowValid(self) -> bool:
        if self.rownum == '' and self.name == '' and self.address == '' and self.area == '':
            return False
        else:
            return True
    
    def getRow(self):
        self.myRow = {'rownum': self.rownum, 'name': self.name, 'address': self.address, 'area': self.area} 
        return self.myRow
            
```

```python
import pdfplumber
import pandas as pd
import numpy as np

def rectifyIt(_df: pd.DataFrame) -> pd.DataFrame:
    df = pd.DataFrame({'rownum':[], 'name': [], 'address': [], 'area': []})
    objRow = clsRow()
    for index, row in _df.iterrows():        
        if len(row) != 4: # one row has 4 columns
            print(index, len(row))
            continue 
        _rownum = row[0].strip()
        _name = row[1].strip()
        _address = row[2].strip()
        _area = row[3].strip()
        if _name == '': # ignore if no name provided           
            continue
        isAdded = objRow.addRow(_rownum, _name, _address, _area)
        # if added successfully and not last row
        if isAdded == True:
            if index == len(_df)-1:
                if objRow.isRowValid() == True:
                    df2 = pd.DataFrame([objRow.getRow()])
                    df = pd.concat([df, df2], ignore_index=True)
                break
            continue
        if objRow.isRowValid() == True:
            new_row = objRow.getRow()
            df_2 = pd.DataFrame([new_row])            
            df = pd.concat([df, df_2], ignore_index=True)    
        objRow.clearIt()
        # add the row again for next use after clearing it
        objRow.addRow(_rownum, _name, _address, _area)
            
    return df
```

```python
with pdfplumber.open('..//Material04//full_list_chi_1.pdf') as pdf:
    page = pdf.pages[0]    
    # table = page.extract_table(table_settings={"vertical_strategy": "lines", 
    #                                            "horizontal_strategy": "text", 
    #                                            "snap_tolerance": 4,})
    mySetting = {"vertical_strategy": "text", "horizontal_strategy": "text"}
    table = page.extract_table(mySetting)
    df = pd.DataFrame(table[::])
    df_1 = rectifyIt(df)
    # print(df_1)
    df_1.to_excel("..\\output04\\doctorPdfList_1.xlsx")
print("complete(1).")
```
